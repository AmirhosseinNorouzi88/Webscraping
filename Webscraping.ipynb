```python
import requests

from bs4 import BeautifulSoup

import sqlite3
```

```python
url = "https://www.iranjib.ir/showarchive/0/0/"

response = requests.get(url)

soap = BeautifulSoup(response.text, "html.parser")
```

```python
div = soap.find("div" , {"class" : "AJAXContent"})
```

```python
table = div.find("table", {"class" : "pagination"})
```

```python
tds = table.find_all("td")
```

```python
tds.pop(-1)
```

```python
urls_archives = []
for td in tds:
    url = td.find("a", {"class" : "jaxlink", "wrapper" : "showarchivetarget"}).get("href")
    urls_archives.append(url)
```

```python
print(urls_archives)
```
